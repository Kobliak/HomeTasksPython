Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents_ = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories_ = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
} 
#ADD DOC FUNCTION==============================================================================================================
def add_doc (directories, documents):
    """
    Function takes two arguments - list of documents and dict which contains list of documents on each shelf
    Function adds document on the shelf
    """
    doc_num = input('Введите номер документа: ')
    shelf_num = input('Введите новый номер полки: ')
    flag = 0
    new_doc = {}
    exist_shelf = ''
    if shelf_num not in directories.keys():
        print(f'Полка №{shelf_num} не существует. Уточните номер')
        return 2
    else:
        for document in documents:
            if doc_num in document.values():
                for directory, docs_on_shelf in directories.items():
                    if doc_num in docs_on_shelf:
                        exist_shelf = directory
                        break
                print(f'Документ №{doc_num} находится на полке №{exist_shelf}')
                flag = 1
                return 0
        if flag != 1:
            new_doc['type'] = input('Введите тип документа: ')
            new_doc['number'] = doc_num
            new_doc['name'] = input('Введите имя: ')
            documents.append(new_doc)
            directories[shelf_num].append(doc_num)
            print(f'Документ №{doc_num} добавлен на полку №{shelf_num}')
            return 1
            
#MOVE FUNCTION=================================================================================================================
def move_doc(directories):
    doc_num = input('Введите номер документа: ')
    shelf_num = input('Введите новый номер полки: ')
    flag = 0
    if shelf_num not in directories.keys():
        print(f'Полки с №{shelf_num} не существует. Уточните номер')
        return 2
    else:
        for shelf, doc_values in directories.items():
            if doc_num in doc_values:
                if shelf == shelf_num:
                    print(f'Документ №{doc_num} уже находится на полке {shelf}.')
                else:
                    print(f'Перемещаю документ №{doc_num} с полки №{shelf} на полку №{shelf_num}')
                    doc_values.remove(doc_num)
                    directories[shelf_num].append(doc_num)
                    flag = 1
                    return 1
        if flag == 0:
            print('Документ с таким номером отсутствует. Уточните номер документа.')
            return 0

#People FUNCTION==============================================================================================================
def people_ (document_list):
    doc_num = input('Введите номер документа или пустую строку для выхода ')
    mark = 0
    if doc_num == '':
        return 2
    for data in document_list:
        if doc_num == data['number']:
            mark = 1
            print('Имя владельца документа:',data['name'])
            return 1
    if mark == 0:
        print('Документ с таким номером не обнаружен. Проверьте правильность номера и повторите ввод. ')
        return 0

#DELETE_DOC FUNCTION==========================================================================================================
def delete_doc (directories, documents):
    flag = 0
    doc_num = input('Введите номер документа: ')
    for docs_directories in directories.values():
        if doc_num in docs_directories:
            docs_directories.remove(doc_num)
            break
    for doc in documents:
        if doc_num == doc['number']:
            documents.remove(doc)
            flag = 1
            break
    if flag == 1:
        print(f'Удаляю документ {doc_num}')
        return 1
    else:
        print('Документ отсутствует. Уточните номер документа')
        return 0

#Shelf for doc FUNCTION======================================================================================================        
def shelf_doc (directories_list):
    doc_num = input('Введите номер документа: ')
    mark = 0
    for shelf_num, shelf_doc in directories_list.items():
        if doc_num in shelf_doc:
            mark = 1
            print(f'Документ находится на полке №{shelf_num}')
            return 1
    if mark == 0:
        print('Документ с таким номером не обнаружен')
        return 0
#Documents List Function=================================================================================    
def list_doc (documents_list):
    output = []
    print('В наличии следующие документы: ')
    for data in documents_list:
        for data_print in data.values():
            output.append(data_print)
        print(', '.join(output))
        output = []
    return 0
#Shelf insert Function======================================================================================    
def shelf_ins (directories):
    new_shelf = input('Введите номер полки: ')
    if new_shelf in directories.keys():
        print('Полка с таким номером уже существует!')
        return 0
    else:
        print(f'Добавляю полку №{new_shelf}')
        directories[new_shelf] = []
        return 1
#List of shelfs============================================================================================
def list_shelf(directories):
    print('Существуют следующие полки: ')
    for shelfs in directories.keys():
        print(f'Полка {shelfs}')
    return 0
```

```python
def main ():
    flag = 0
    check = ''
    while True:
        command = input('Введите команду. Exit для выхода: ')
        if command.lower() != 'exit':
            if command.lower() in ['p', 'people']:
                flag = people_(documents_)
                while flag == 0:
                    flag = people_(documents_)
            elif command.lower() in ['s', 'shelf']:
                flag = shelf_doc (directories_)
                while flag == 0:
                    check = input('Желаете ввести новый номер документа? Y/N: ')
                    if check.lower() == 'y':
                        flag = shelf_doc (directories_)
                    else:
                        break
            elif command.lower() in ['l', 'list']:
                list_doc(documents_)
            elif command.lower() in ['as', 'add shelf']:
                flag = shelf_ins(directories_)
                while flag == 0:
                    check = input('Желаете ввести новый номер полки? Y/N: ')
                    if check.lower() == 'y':
                        flag = shelf_ins(directories_)
                    else:
                        break
            elif command.lower() in ['d', 'delete']:
                flag = delete_doc(directories_, documents_)
                while flag == 0:
                    check = input('Желаете ввести новый номер документа? Y/N: ')
                    if check.lower() == 'y':
                        flag = delete_doc(directories_, documents_)
                    else:
                        break
            elif command.lower() in ['m', 'move']:
                flag = move_doc(directories_)
                while flag !=1:
                    check = input('Желаете повторить ввод? Y/N: ')
                    if check.lower() == 'y':
                        flag = move_doc(directories_)
                    else:
                        break
            elif command.lower() in ['a', 'add']:
                flag = add_doc(directories_, documents_)
                while flag !=1:
                    check = input('Желаете повторить ввод? Y/N: ')
                    if check.lower() == 'y':
                        flag = add_doc(directories_, documents_)
                    else:
                        break
            elif command.lower() in ['ls', 'list shelf']:
                list_shelf(directories_)
            else:
                continue
        else:
            print('Exit')
            break    
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
*Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ*.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. *Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. *Предусмотрите сценарий, когда пользователь вводит несуществующий документ*;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. *Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку*.
