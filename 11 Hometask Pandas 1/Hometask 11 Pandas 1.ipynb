```python
#Task2
data = pd.read_csv('power.csv')
filtered = data[
                ((data['country'] == 'Estonia') | (data['country'] == 'Latvia') | (data['country'] == 'Lithuania')) & 
                (data['year'] < 2011) & (data['year'] > 2004) &
                ((data['category'] == 4) | (data['category'] == 12) | (data['category'] == 21))  
                & (data['quantity'] > 0 )
               ]
print(filtered['quantity'].sum())
filtered.shape
```
