```python
#Task 1
def classify (rating):
    if rating <= 2.0:
        return ('Low')
    elif rating <= 4.0:
        return ('Medium')
    else:
        return ('High')  

ratings = pd.read_csv('ratings.csv')
ratings['class'] = ratings['rating'].apply(classify)
ratings.head(15)
```

```python
#Task 2
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

def region_fill(row):
    for word in row['split']:
        if word.lower() in geo_data['Центр']:
            return 'Центр'
        elif word.lower() in geo_data['Северо-Запад']:
            return 'Северо-Запад'
        elif word.lower() in geo_data['Дальний Восток']:
            return 'Дальний Восток'
        else:
            continue
    return 'undefined'

keywords = pd.read_csv('keywords.csv')
keywords['split'] = keywords['keyword'].apply(lambda phrase: phrase.split(' '))
keywords['region'] = keywords.apply(region_fill, axis=1)
del keywords['split']

keywords['region'].unique()
```
