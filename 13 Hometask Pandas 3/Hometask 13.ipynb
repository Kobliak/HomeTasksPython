```python
import pandas as pd
```

```python
#Task 1

log = pd.read_csv('visit_log.csv', sep=';')
traf_1 = ['yandex', 'google']                              #Первый вид трафика
traf_2 = ['paid', 'email']                                 #Второй вид трафика

log.loc[log.traffic_source.isin(traf_1), 'source_type'] = 'organic' #Заполняем столбец source_type для яндекса и гугла
log.loc[log.traffic_source.isin(traf_2), 'source_type'] = 'other' #Заполняем столбец source_type для всех строк paid и email
log.loc[(log.traffic_source.isin(traf_2)) & (log.region == 'Russia'), 'source_type'] = 'ad' 
                                                     #Заполняем столбец source_type для всех строк paid и emai, но для России
log.loc[pd.isna(log.source_type), 'source_type'] = log.traffic_source #Все остальные строки заполняем значениями traffic_source

log.head()
```

```python
#Task 3
from pymystem3 import Mystem
m = Mystem()

data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})


def lemma(row):
    lemmas = m.lemmatize(row.keyword)
    return (str(' '.join(lemmas)))    

data['lemmas'] = data.apply(lemma, axis=1)
```
