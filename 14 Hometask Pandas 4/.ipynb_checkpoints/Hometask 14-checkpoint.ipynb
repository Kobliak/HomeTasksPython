```python
import pandas as pd
```

```python
#Task 1
ratings = pd.read_csv('ratings.csv', encoding = 'utf8')
asd = ratings.groupby('userId').count()
user_100 = asd.loc[asd.movieId > 100]
del user_100['rating']
del user_100['timestamp']
user_100.columns = ['Qty']
new_df = ratings.merge(user_100, how='right', on='userId')
new_df1 = new_df.groupby('userId').agg(['min', 'max'])['timestamp']
av_time = int((new_df1['max'] - new_df1['min']).mean())
av_time
```

Для связывания данных координат и сведений о визитах и покупках я бы использовал данные timestamp определения/взятия координатах и данные timestamp совершенных визитов и покупок, при условии, что такие данные доступны.

timestamp координат и визитов/покупок случае получения данных из разных систем могут различаться и не совпадать. В этом случае будет необходимо для привязки координат и визитов/покупок использовать некоторую минимально допустимую разницу для определения соответствия набора координат факту покупки. Т.е. если координаты получены за небольшой промежуток времени, то возможно, что их можно объединить в один факт и искать соответствие визита/покупки тоже с небольшой дельтой относительно timestamp визита/покупки

Факт значительного удаления координат относительно друг друга за короткий промежуток времени может быть, как ошибками данных (подозрительный выброс), так и возможностью использования аккаунта для покупок разными людьми, например, членами одной семьи. И если выбросы необходимо удалять из обработки, то факт использования аккаунта разными людьми должна соответствовать правилам пользования приложения. Такой факт требует специальной обработки.

Из координат можно получить географическое нахождение пользователей приложения по территории страны на основании распределения координат по карте местности для предложения региональных товаров и акций; частоту, продолжительность и расстояния перемещения пользователей приложения для анализа возможностей предложения соответствующих товаров для туристов, командировочных и пр.
