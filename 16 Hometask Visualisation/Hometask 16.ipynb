```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
player_data = pd.read_csv('player_data.csv')
```

```python
#Firsr visualisation
# This visualisation shows  NBA players distribution for each University
# Output - if you want to be NBA player you shoud graduate University of Kentucky
pl2 = pd.DataFrame(player_data.college.value_counts().reset_index())
pl2.columns = ['college','Qty_players']
plt.barh(pl2['college'].head(10), pl2['Qty_players'].head(10))
plt.xlabel('Quantity')
plt.ylabel('University')
plt.title('Top 10 Universities with NBA players')
```

```python
# Second visualisation
# This visualisation shows the players distibution for each position
# Outout - guard are the most numerous plaers among NBA players and Center-forwards are the smallest by quantity 
pos_stats = pd.DataFrame(player_data.groupby(['position']).position.value_counts().sort_values(ascending=False))
pos_stats.columns = ['Qty_players']
pos_stats.plot(kind='pie', y='Qty_players')
```

```python
players = pd.read_csv('Players.csv')
players.head(15)
```

```python
# Third visualisation
# Visualisation shows players height distribution
# 206 cm is the the most numerous height of the NBA players
ax = players.height.hist(bins=25)
ax.set_title('Players height distribution')
ax.set_xlabel('Players height')
ax.set_ylabel('Players')
```

```python
Seasons_Stats = pd.read_csv('Seasons_Stats.csv')
```

```python
# Forth visualisation
# This visualisation shows the points distribution for each position 
Points = Seasons_Stats.Pos.value_counts().sort_values(ascending = False).head(7).index.values
ax = sns.boxplot(x="Pos", y="PTS", 
            data=Seasons_Stats[Seasons_Stats.Pos.isin(Points)])
```

```python
# Fifth visualization
# visualization shows correlation for years. points for each year 
# Output - mid of the 60ths and 90ths are stands out from average season total points.
sns_plot = sns.pairplot(
    Seasons_Stats[['Pos', 'Year', 'PTS']])
sns_plot
```
