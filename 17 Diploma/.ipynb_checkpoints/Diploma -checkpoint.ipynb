```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Part 1
hr = pd.read_csv('hr.csv')
hr.head(10)
```

```python
#Part 2 ----------------------
corr = hr.corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(3)
```

```python
# Part 6------------------------------------------------------------
employee_qty = hr.groupby('sales').count().reset_index()
employee_qty = employee_qty.iloc[:,0:2]
employee_qty.columns = ['sales', 'employee_qty']
employee_qty.head(10)
```

```python
#Part 7 -------------------------
salary_dist = hr.groupby('salary').count()#.reset_index()
salary_dist = salary_dist.iloc[:,0:2]
salary_dist.columns = ['sales', 'salary_qty']
#salary_dist.head()
salary_dist.plot(kind='pie', y='salary_qty')
```
