```python
hr_1 = hr.copy()

#приведем нечисловые столбцы к числовым значениям
level_map1 = {'sales': 1, 'accounting': 2, 'hr': 3, 'technical': 4, 'support': 5, 'management': 6,
       'IT': 7, 'product_mng': 8, 'marketing': 9, 'RandD': 10}
hr_1['sales_d'] = hr['sales'].map(level_map1)
level_map = {'high': 3, 'medium':2, 'low': 1}
hr_1['salary_d'] = hr['salary'].map(level_map)

del hr_1['sales']
del hr_1['salary']
hr_1.head()
```

*** Гипотеза 1: средние выборок по работникам (строкам) равны

```python
T_T_1 = []
for _ in range(1000):
    index_1 = random.randint(0, int((len(hr_1))**0.5))
    index_2 = random.randint(0, int((len(hr_1))**0.5))
    index_3 = random.randint(0, int((len(hr_1))**0.5))
    index_4 = random.randint(0, int((len(hr_1))**0.5))
    data1 = hr_1.iloc[index_1 * index_2].values.tolist()
    data2 = hr_1.iloc[index_3 * index_4].values.tolist()
    T_T_1.append(stats.ttest_ind(data1, data2).pvalue)
print(np.mean(T_T_1))
```

** Гипотеза 2: средние по выборкам среднемесячных часов работы по разным отделам равны

```python
rt_1 = hr_1.loc[hr_1['sales_d'] == 1]
rt_2 = hr_1.loc[hr_1['sales_d'] == 4]
sam8 = rt_1.loc[:850, 'average_montly_hours']
sam9 = rt_2.loc[:850, 'average_montly_hours']
t_test2 =stats.ttest_ind(sam8, sam9)
t_test2
```

```python
** Гипотеза 3: средние по выборкам удовлетворенности сотрудников с разным уровнем заплаты равны
```

```python
rt_4 = hr_1.loc[hr_1['salary_d'] == 1]
rt_5 = hr_1.loc[hr_1['salary_d'] == 2]
sam5 = rt_4.loc[:5000, 'satisfaction_level']
sam6 = rt_5.loc[:5000, 'satisfaction_level']
t_test3 =stats.ttest_ind(sam5, sam6)
t_test3
```

** pvalue <0.05, следовательно гипотезу нужно отвегнуть
